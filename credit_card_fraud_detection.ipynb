```python
#build a sample dataset containing similar distribution of normal and fraudalent transactions so basically we are picking 492 fraudalent and 492 normal transaction

normal_sample = normal.sample(n=492)
```

```python
print(normal_sample) #these are random values.
```

```python
#concatinating the two dataframes normal and fraud  here axis = 0 means concatinating by rows axis = 1 means concatinating by cols

newdata = pd.concat([normal_sample,fraud],axis=0)
```

```python
print(newdata)
```

```python
newdata.head()
```

```python
newdata.tail()
```

```python
newdata.groupby('Class').mean()
```

```python
#splitting data into feature and target

x = newdata.drop(columns='Class',axis=1)
y = newdata['Class']
```

```python
print(x)
print(y)
```

```python

#split data into training data and testing data

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,stratify=y,random_state=2)
```

```python
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
#logistic regression model

model = LogisticRegression()
```

```python
#training the model with training data

model.fit(x_train,y_train)

#X_train contains of all the fetures of training data

#y_train contains of label
```

```python
#model evaluation by accuracy score

#accuracy on training data
x_train_pred = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_pred,y_train)
```

```python
print('accuracy on training data:',training_data_accuracy)  #94.186 pecent it made correct predictions with comparing original label and trained data.

```

```python
#accuracy on test data

x_test_pred = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_pred,y_test)
```

```python
print('accuracy on test data:',test_data_accuracy) #on testing it made 91.55 percent correct predictions with comparing testing data and original data


#so conclusion is our model has performed well on both training and test data.
```
