```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

#importing the dataset

```python
dataset = pd.read_csv('bigbasket.csv',header=None)
```

```python
print(dataset)
```

```python
transactions = dataset.iloc[:7219, :20].astype(str).values.tolist()
```

transactions in 2d array

```python
print(transactions)
```

training model using apriori algorithm

```python
from apyori import apriori
basket_intelligence = apriori(transactions=transactions, min_support = 0.003,min_confidence=0.2,min_lift = 3,min_len = 2)
```

#visualizing

raw results

```python
results = list(basket_intelligence)
results
```

proper display

```python
def inspect(results):
  product1 = [tuple(result[0])[0] for result in results if result[0]]
  product2 = [tuple(result[0])[1] if len(result[0]) > 1 else 'N/A' for result in results]

  supports = [result[1] for result in results]

  confidences = [result[2][0][2] for result in results]
  lifts = [result[2][0][3] for result in results]
  return pd.DataFrame(list(zip(product1,product2,supports,confidences,lifts)),columns=['product1','product2','support','confidence','lift'])
```

```python
Dataframe_intelligence = inspect(results)
display(Dataframe_intelligence)
```

```python
Dataframe_intelligence.nlargest(n=10,columns='lift')
```
